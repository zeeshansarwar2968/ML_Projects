## Seaborn Bar plot
Question 1
> Write a program to draw bar plot of sex against survived for a dataset given in the url
[https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv]


```python
sns.set(style='whitegrid')
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")
print(df.head())
sns.barplot(x='sex', y='survived', data=df)
```


## Plotly Scatterplot


Question 1
> Write a program to draw a scatter plot for random 1000 x and y coordinates
> 


```python
# np.random.seed(42) 
    
rand_x= np.random.randint(1,1001,1000) 
rand_y= np.random.randint(1,1001,1000) 
    
plot = px.scatter(rand_x, rand_y)
plot.show()
```

Question 2
> Write a program to draw line and scatter plots for random 100 x and y coordinates
> 


```python
rand_x= np.random.randint(1,101,100) 
rand_y= np.random.randint(1,101,100) 
    
plot1 = px.scatter(rand_x, rand_y)
plot2 = px.line(rand_x, rand_y)

plot1.show()
plot2.show()
```

Question 3
> Write a program to draw a scatter plot for random 500 x and y coordinates and style it



```python
    
rand_x= np.random.randint(1,501,500) 
rand_y= np.random.randint(1,501,500) 
    
plot = px.scatter(x=rand_x,y=rand_y, color=rand_y, size=rand_x)
plot.show()
```

Question 4
> Write a program to draw a scatter plot for a given dataset and show datalabels on hover
https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv")
print(df.head(20))
plot = px.scatter(df,x='State', y='Population', color="State", hover_data=list(df.columns))
plot.show()
```
